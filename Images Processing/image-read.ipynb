```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
root_path = '/kaggle/working/'
path = '/kaggle/input/images-of-bird/bird-8788491_1280.jpg'
```

```python
def read_and_save_images(image_path, text_file_path):
  img = Image.open(image_path)

  plt.imshow(img)
  # plt.axis('off')
  plt.show()

  output_images_file = root_path + 'original_output.jpg'
  img.save(output_images_file)

  if os.path.exists(text_file_path):
    os.remove(text_file_path)
  pixel_value = list(img.getdata())

  with open(text_file_path, 'w') as f:
    for pixel in pixel_value:
      f.write(f"{pixel[0]}, {pixel[1]}, {pixel[2]}\n")



input_file = path
output_file = root_path + 'RGB_text_file.txt'
read_and_save_images(input_file, output_file)
```

**Convert Into Gray Scale**

```python
def rgb_to_grayscale(image_path, output_path):
  img = Image.open(image_path)
  plt.imshow(img)
  plt.axis('off')
  plt.show()

  gray_img = img.convert('L')       #Grayscale value = 0.299 * Red + 0.587 * Green + 0.114 * Blue
  plt.imshow(gray_img, cmap='gray')
  plt.axis('off')
  plt.show()
  img.save(output_path)


input_file = path
gray_img_output_file = root_path + 'grayscale_output.jpg'
rgb_to_grayscale(input_file, gray_img_output_file)
```

```python
def grayscale_to_txt(image_path, text_file_path):
  img = Image.open(image_path)
  if os.path.exists(text_file_path):
    os.remove(text_file_path)
  pixel_value = list(img.getdata())

  with open(text_file_path, 'w') as f:
    for pixel in pixel_value:
      f.write(f"{pixel}\n")

input_file = gray_img_output_file
gray_txt_output_file = root_path + 'grayscale_text_file.txt'
grayscale_to_txt(input_file, gray_txt_output_file)
```
